```python
import numpy as np
import pandas as pd
import keras as kr
from keras.models import Sequential
from keras.layers import Dense

dataset = pd.read_csv('3-cleaned-normalized.csv')
dataset.head()
```

```python
size = dataset.shape[0]
```

```python
X = dataset[['gender', 'smoke', 'alco', 'active', 
             'age', 'height', 'weight', 'ap_hi', 'ap_lo',
            'gluc', 'cholesterol']].values
```

```python
Y = dataset['cardio'].values
```

```python
model = Sequential()
model.add(Dense(15, input_dim=11, activation='tanh'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(1, input_dim=11, activation='sigmoid'))
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, Y, epochs=200, batch_size=10)
```

```python
loss, accuracy = model.evaluate(X, Y)
print("\nLoss: %.2f, Accuracy: %.2f%%" % (loss, accuracy*100))
```

```python
probabilities = model.predict(X)
predictions = [float(round(x[0])) for x in probabilities]
accuracy = np.mean(predictions == Y)
print("Prediction Accuracy: %.2f%%" % (accuracy*100))
```
